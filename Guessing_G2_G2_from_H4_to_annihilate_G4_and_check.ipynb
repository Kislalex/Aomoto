```python
test = ['l_1','l_2','l_3','l_4','m_1','m_2','m_3','m_4']
q_,k_ = Alt(test)
res = []
for q,k in list(zip(q_,k_)):
    if 'l' in q[7]:
        continue
    #print(q[7])
    base = [
        Delta(['l_0','m_0',q[0],q[1],q[2]]),
        Delta(['l_0','m_0',q[0],q[1],q[3]])
    ]
    base[0].sort()
    base[1].sort()
    dt = 1 - 2 * mySort(base)
    g_1 = GGroup(['l_0','m_0',q[2]],[q[0],q[4],q[5],q[6]])
    k_1 = -2 * g_1.sort()
    g_2 = GGroup(['l_0','m_0',q[0]],[q[2],q[4],q[5],q[6]])
    k_2 = 0*1000000 * g_2.sort()
    g_3 = GGroup(['l_0','m_0',q[0]],[q[1],q[4],q[5],q[6]])
    k_3 = 1 * g_3.sort()
    res.append((copy.deepcopy(base) + [g_1], k_1 * k * dt))
    res.append((copy.deepcopy(base) + [g_2], k_2 * k * dt))
    res.append((copy.deepcopy(base) + [g_3], k_3 * k * dt))
    
reduction(res)
```
